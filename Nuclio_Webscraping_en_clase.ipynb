### Marca

```python
response = requests.get("https://www.marca.com/")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
tabla = soup.find_all("div", class_="main-nav-container")
tabla
```

```python
lista = []
for i in tabla:
  lista.append((i.text).replace("\n", ","))
```

```python
lista
```

```python
l = str(lista) #.replace(",,,,,", ",")
l = l.strip("[', ]")
l
```

```python
l = l.split(",")
l
```

```python
nom = []
for i in l:
  if len(i) == 0:
    pass
  else:
    nom.append(i)
nom
```

```python
imagen = soup.find_all("img")
len(imagen)
```

```python
imagen
```

```python
tags = []
for i in imagen:
  if "src" in i.attrs:
    if i.attrs["src"].find("soccer")>0:
      tags.append(i.attrs["src"])
```

```python
tags, len(tags)
```

```python
for i in tags:
  fichero = open("/content/drive/MyDrive/DATA SCIENCE/Nuclio DS/Web Scrappping/Webscraping/escudos/"+basename(i), "wb")
  fichero.write(requests.get(i).content)
```
